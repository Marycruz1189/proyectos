### Feature Selection ó Selección de Características

Para mejorar nuestros resultados con este algoritmo.En vez de utilizar las 11 columnas de datos de entrada que tenemos, vamos a utilizar una Clase de
SkLearn llamada SelectKBest con la que seleccionaremos las 5 mejores características y usaremos
sólo esas.

```python
X=dataframe.drop(['comprar'], axis=1)
y=dataframe['comprar']

best=SelectKBest(k=5)
X_new = best.fit_transform(X, y)
X_new.shape
selected = best.get_support(indices=True)
print(X.columns[selected])
```

Veamos qué grado de correlación tienen:

```python
used_features =X.columns[selected]

colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(dataframe[used_features].astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

### Crear el modelo Gaussian Naive Bayes con SKLearn

```python
# Split dataset in training and test datasets
X_train, X_test = train_test_split(dataframe, test_size=0.2, random_state=6)
y_train =X_train["comprar"]
y_test = X_test["comprar"]
```

```python
# Instantiate the classifier
gnb = GaussianNB()
# Train classifier
gnb.fit(
X_train[used_features].values,
y_train
)
y_pred = gnb.predict(X_test[used_features])

print('Precisión en el set de Entrenamiento: {:.2f}'.format(gnb.score(X_train[used_features], y_train)))
print('Precisión en el set de Test: {:.2f}'.format(gnb.score(X_test[used_features], y_test)))
```

### Probemos el modelo: ¿Comprar o Alquilar?

```python
# ['ingresos', 'ahorros', 'hijos', 'trabajo', 'financiar']
print(gnb.predict([[2000, 5000, 0, 5, 200000],[6000, 34000, 2, 5, 320000] ]))
#Resultado esperado 0-Alquilar, 1-Comprar casa
```
